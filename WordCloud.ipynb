```python
wc = WordCloud(stopwords=STOPWORDS,background_color='white')
wc.generate(text)
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(wc)
```

```python
sw =set(STOPWORDS)
sw.add('Mr')
```
